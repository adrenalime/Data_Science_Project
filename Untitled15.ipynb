```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from matplotlib.pyplot import figure
%matplotlib inline
df1 = pd.read_csv("2020.csv")
df2 = pd.read_csv("2021.csv")
df3 = pd.read_csv("WHR20_DataForTable2.1.csv")
#df = df[['Ladder score','Logged GDP per capita','Social support','Healthy life expectancy','Freedom to make life choices','Generosity','Perceptions of corruption']]

```

```python
df1=df1[df1["Regional indicator"] == "Western Europe"]
df2=df2[df2["Regional indicator"] == "Western Europe"]
df3=df3[df3["Regional indicator"] == "Western Europe"]
```

```python
fig, ax = plt.subplots()

ax.bar(df2['Country name'], df2['Ladder score'])
plt.xticks(rotation=30, ha='right')
fig.set_size_inches(12, 8, forward=True)
plt.title('The relationship between Generosity and Happiness in Western Europe')
plt.xlabel('Country name')
plt.ylabel('Happiness Score')

x=df2['Country name']
y=df2['Generosity']
ax2 = ax.twinx()
ax2 = plt.plot(x, y,color ='r')
plt.ylabel('Generosity',color = 'r')
plt.show()
```

```python
df5 = pd.read_csv("2021.csv")
df5 = df5.head(20)
bestwe = df5[df5["Regional indicator"] == "Western Europe"]
bestna= df5[df5["Regional indicator"] == "North America and ANZ"]
bestme= df5[df5["Regional indicator"] == "Middle East and North Africa"]
bestla= df5[df5["Regional indicator"] == "Latin America and Caribbean"]
bestcee= df5[df5["Regional indicator"] == "Central and Eastern Europe"]
bestsa= df5[df5["Regional indicator"] == "Southeast Asia"]
bestcoi= df5[df5["Regional indicator"] == "Commonwealth of Independent States"]
bestssa= df5[df5["Regional indicator"] == "Sub-Saharan Africa"]
bestea= df5[df5["Regional indicator"] == "East Asia"]
```

```python
bestwe_counted= len(bestwe.index)
bestna_counted= len(bestna.index)
bestme_counted= len(bestme.index)
bestla_counted= len(bestla.index)
bestcee_counted= len(bestcee.index)


x= [bestwe_counted,bestna_counted,bestme_counted,bestla_counted,bestcee_counted]
labels = ['Western Europe', 'North America and ANZ', 'Middle East and North Africa', 'Latin America and Caribbean','Central and Eastern Europe']
plt.title("The most happiest 20 countries in the world")
plt.pie(x, startangle = 90,labels=labels)

centre_circle = plt.Circle((0,0),0.50,fc='white')
fig = plt.gcf()
fig.gca().add_artist(centre_circle)
plt.show() 
```
